#### Seaborn - Built on top of matplotlib

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset("tips")

# Scatter plot of total_bill vs tip
sns.scatterplot(x="total_bill", y="tip", data=tips, hue="sex", style="time")
plt.title("Scatter Plot Example")
plt.show()
```

```python
sns.barplot(x="day", y="total_bill", data=tips, hue="sex")
plt.title("Bar Plot Example")
plt.show()
```

```python
sns.histplot(tips["total_bill"], bins=30, kde=True, color="blue") # for density curve ! 
plt.title("Histogram Example")
plt.show()
```

```python
sns.boxplot(x="day", y="total_bill", data=tips, hue="sex") # good for outliers ! 
plt.title("Box Plot Example")
plt.show()
```

```python
sns.violinplot(x="day", y="total_bill", data=tips, hue="sex", split=True) # Combines box plot and KDE plot for detailed distribution visualization.
plt.title("Violin Plot Example")
plt.show()
```

```python
sns.pairplot(tips, hue="sex") # pairs 
plt.show()


# for each feature: 
#       compare the 2 independent variable
#               plot density comparison 
#               plot scatter plot of the SAME variable, for the different indepedenent var 
#                NOTE THE AXIS ! 
```

```python
sns.regplot(x="total_bill", y="tip", data=tips) # regression line 
plt.title("Regression Plot Example")
plt.show()
```

```python
sns.countplot(x="day", data=tips, hue="sex") # count the number of categorical instances 
plt.title("Count Plot Example")
plt.show()
```

```python
sns.jointplot(x="total_bill", y="tip", data=tips, kind="scatter") # combines scatter + histogram 
plt.show()
```

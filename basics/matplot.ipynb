### Matplot Library 

The basics of visualizations 

1. Standard Plot - Time Series 

2. Scatter Plot - Relationship between 2 variable 

3. Bar Chart - For categorical 

4. Historgram - For frequency  

5. Box Plot - Distributions + Outliers 

6. Correlation Matrix - Heat map 

7. Pie Chart - Proportions

8. Subplots - Multiple graphs in One Figure

Reference: https://matplotlib.org/cheatsheets/cheatsheets.pdf

```python
import matplotlib.pyplot as plt
import numpy as np 
import seaborn as sns # fancier data visulization based on matplotlib 
import pandas as pd 
```

```python
# Line plot 

x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 40]

plt.plot(x, y, marker='o', linestyle='-', color='b', label="Line")  
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Line Plot Example")
plt.legend()
plt.show()
```

```python
# scatter plot 

x = np.random.rand(50)
y = np.random.rand(50)

plt.scatter(x, y, color='r', alpha=0.5)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Scatter Plot Example")
plt.show()
```

```python
# bar chart 

categories = ['A', 'B', 'C', 'D']
values = [10, 20, 15, 25]

plt.bar(categories, values, color=['blue', 'green', 'red', 'orange'])
plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Bar Chart Example")
plt.show()
```

```python
# horizontal bar chart 

plt.barh(categories, values, color=['purple', 'cyan', 'yellow', 'pink'])
plt.xlabel("Values")
plt.ylabel("Categories")
plt.title("Horizontal Bar Chart Example")
plt.show()
```

```python
# Histogram 

data = np.random.randn(1000)

plt.hist(data, bins=30, color='g', alpha=0.7, edgecolor='black')
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Histogram Example")
plt.show()
```

```python
# Box plot 

data = [np.random.randn(100), np.random.randn(100), np.random.randn(100)]

plt.boxplot(data, labels=['Set 1', 'Set 2', 'Set 3'])
plt.ylabel("Values")
plt.title("Box Plot Example")
plt.show()
```

```python
# Correlation map 

data = {
    "A": np.random.rand(50),
    "B": np.random.rand(50),
    "C": np.random.rand(50),
    "D": np.random.rand(50)
}

df = pd.DataFrame(data)

# Compute correlation matrix
corr_matrix = df.corr()

# Plot heatmap
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Matrix Heatmap")
plt.show()
```

```python
# Pie chart 

labels = ['Apple', 'Banana', 'Cherry', 'Date']
sizes = [20, 30, 25, 25]
colors = ['blue', 'yellow', 'red', 'green']

plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title("Pie Chart Example")
plt.show()

```

```python
# Sub plots 

fig, axs = plt.subplots(2, 2, figsize=(10, 8)) # the juicer 

# First subplot: Line Plot
axs[0, 0].plot(x, y, color='b', marker='o')
axs[0, 0].set_title("Line Plot")

# Second subplot: Scatter Plot
axs[0, 1].scatter(x, y, color='r', alpha=0.5)
axs[0, 1].set_title("Scatter Plot")

# Third subplot: Bar Chart
axs[1, 0].bar(categories, values, color=['blue', 'green', 'red', 'orange'])
axs[1, 0].set_title("Bar Chart")

# Fourth subplot: Pie Chart
axs[1, 1].pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
axs[1, 1].set_title("Pie Chart")

plt.tight_layout()
plt.show()
```
